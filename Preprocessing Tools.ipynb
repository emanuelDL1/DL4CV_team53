```python
from utils import get_screen,get_screen_resize,rgb2gr,rgb2y,rgb2bw
# Fast functions
from utils import crop, downsample, breakout_preprocess, gray2pytorch, rgb2gray, rgb2pytorch

import time

import gym
import matplotlib.pyplot as plt
```

# Simpler & faster implementation 

```python
env = gym.make('Breakout-v0')

start = time.time()
frame = env.render(mode='rgb_array')
frame = breakout_preprocess(frame)
end = time.time()
print('preprocessing time', end - start)

#env.reset()
plt.figure()
plt.imshow(frame, interpolation='none', cmap='gray')
plt.title('Preprocessed screen')

plt.show()
```

# Old & more exact implementation

```python
#env.reset()
plt.figure()
plt.imshow(get_screen(env).cpu().squeeze(0).permute(1, 2, 0).numpy(),
           interpolation='none')
plt.title('Example extracted screen')

plt.figure()
plt.imshow(get_screen_resize(env).cpu().squeeze(0).permute(1, 2, 0).numpy(),
           interpolation='none')
plt.title('Example extracted & resized screen')

plt.figure()
plt.imshow(rgb2y(get_screen_resize(env)).cpu().squeeze().squeeze(0).numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted & resized lumincance screen')

plt.figure()
plt.imshow(rgb2gr(get_screen_resize(env)).cpu().squeeze().squeeze(0).numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted & resized gray screen')

plt.figure()
plt.imshow(rgb2bw(get_screen_resize(env)).cpu().squeeze().numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted & resized bw screen')

start = time.time()
rgb2gr(get_screen_resize(env))
end = time.time()
print('preprocessing time', end - start)

plt.show()
```
