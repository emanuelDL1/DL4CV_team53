```python
from utils import get_screen,get_screen_resize,rgb2gray,rgb2y,rgb2bw
import gym
import matplotlib.pyplot as plt
```

```python
env = gym.make('Breakout-v0')
```

```python
#env.reset()
plt.figure()
plt.imshow(get_screen(env).cpu().squeeze(0).permute(1, 2, 0).numpy(),
           interpolation='none')
plt.title('Example extracted screen')

plt.figure()
plt.imshow(get_screen_resize(env).cpu().squeeze(0).permute(1, 2, 0).numpy(),
           interpolation='none')
plt.title('Example extracted & resized screen')

plt.figure()
plt.imshow(rgb2y(get_screen_resize(env)).cpu().squeeze().squeeze(0).numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted lumincance screen')

plt.figure()
plt.imshow(rgb2gray(get_screen_resize(env)).cpu().squeeze().squeeze(0).numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted gray screen')

plt.figure()
plt.imshow(rgb2bw(get_screen_resize(env)).cpu().squeeze().numpy(),
           interpolation='none', cmap='gray')
plt.title('Example extracted bw screen')


plt.show()
```
